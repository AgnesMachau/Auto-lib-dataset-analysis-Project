```
import pandas as pd
import numpy as np
import csv
```

Loading the Telcom datasets

```
#loading the first Telcom dataset
df1 = pd.read_csv('Telcom_dataset.csv') 
df1.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B',
               'CELL_ID', 'SITE_ID']
df1
```

```
#loading the second Telcom dataset
df2 = pd.read_csv('Telcom_dataset2.csv')
df2.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B',
               'CELL_ID', 'SITE_ID']
df2
```

```
#loading the third Telcom dataset
df3 = pd.read_csv('Telcom_dataset3.csv')
df3.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B',
               'CELL_ID', 'SITE_ID']
df3
```

**Combining the three telecom datasets**

```
#combining the three datasets
tel_combined = pd.concat([df1, df2, df3], axis = 0)
tel_combined
```

**Telecom data cleaning**

```
#checking for any duplicates
#dropping any duplicates
tel_combined.duplicated()
tel_combined1 = tel_combined.drop_duplicates()
tel_combined1
```

```
#checking for any missing values
#dropping any missing values
tel_combined1.isnull()
tel_combined2 = tel_combined1.dropna()
tel_combined2
```

```
#droping columns that are not useful
tel_combined3 = tel_combined2.drop(columns = ['CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'SITE_ID'])
tel_combined3
```

```
#dropping rows which have all values as NA
tel_combined4 = tel_combined3.dropna(how = 'all')
tel_combined4
```

```
#separating date and time
tel_combined4[['DATE', 'TIME']] = tel_combined4.DATE_TIME.str.split(expand = True)
tel_combined4.head()
```

**Loading the cells_geo dataset**

```
#Loading a cell_geo csv
df4 = pd.read_csv("cells_geo.csv", delimiter =";")
df4.columns = ['INDEX', 'CITY', 'STATUS', 'IN_ABIDJAN_OR_NOT', 'GEO_ZONE', 'ZONE_NAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'CELL_ID', 'SITE_ID']
df4.head()
```

```
#loading the cells_geo_description data
df5 = pd.read_excel("cells_geo_description.xlsx")

df5.head(10)
```

```
#loading the CDR_description data
df6 = pd.read_excel("CDR_description.xlsx")

df6.head(10)
```

**Cells_geo data cleaning**

```
#checking if there are any duplicates
#droppind any duplicates
df4.duplicated()
cell_geo = df4.drop_duplicates()
cell_geo
```

```
#checking for any missing values
#dropping any missing values
cell_geo.isnull()
cell_geo1 = cell_geo.dropna()
cell_geo1
```

```
#dropping rows whose status are not 'in service'
cell_geo2 = cell_geo1[cell_geo1['STATUS'] != 'IN SERVICE'].dropna()
cell_geo2
```

```
#dropping columns that are not useful
cell_geo3 = cell_geo2.drop(columns = ['INDEX', 'STATUS', 'GEO_ZONE', 'ZONE_NAME', 'LONGITUDE', 'LATITUDE', 'AREA', 'SITE_ID'])
cell_geo3
```

```
#dropping rows with all values as NA
cell_geo4 = cell_geo3.dropna(how = 'all')
cell_geo4
```

```
#combining the two datasets (telcom dateset and cells_geo dataset)

dataset = tel_combined4.merge(cell_geo4, on = ['CELL_ID'], how = 'outer')
dataset
```

```
#informatin of the combined dataset
print(dataset.info())
```

**Analysis**

```
#finding the product that was used the most
usage = dataset.groupby(by = 'PRODUCT').agg('count').reset_index()
usage
```

```
#the product with the most value
value_product = dataset.groupby(['PRODUCT'])[['VALUE']].sum().reset_index()
value_product
```

```
value_product.plot.bar(x = 'PRODUCT', y = 'VALUE')
```

```
#which is the most used city for the three days
pd.set_option('display.max_rows', None)
most_used_city = dataset['CITY'].value_counts()
most_used_city.head()
```

```
#whether MTNservices were mostly used in ABIDJAN or outside ABIDJAN
Abidjan_used = dataset['IN_ABIDJAN_OR_NOT'].value_counts()
Abidjan_used
```

```
#regions that used MTN services the most
most_region = dataset['REGION'].value_counts()
most_region
```
